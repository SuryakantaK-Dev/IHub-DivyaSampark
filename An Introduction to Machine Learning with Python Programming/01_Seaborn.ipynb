# distplot

shows distribution of a univariate(one variable) set of observations. 

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
sns.distplot(df['total_bill'])
```

kde: kernel density estimation calculates the density of features in the neighbourhood around those features. 

```python
sns.distplot(df['total_bill'],kde=True, bins = 30)

#bins define smoothness or sharpness in your plot
```

# jointplot

match up two distplots for bivariate data. kind parameter: 'scatter', 'reg','kde','hex','resid'

```python
sns.jointplot(x='total_bill',y='tip',data=df, kind='scatter')
plt.show()
```

```python
sns.jointplot(x='total_bill',y='tip',data=df, kind='hex')
plt.show()
```

```python
sns.jointplot(x='total_bill',y='tip',data=df, kind='reg')
plt.show()
```

```python
sns.jointplot(x='total_bill',y='tip',data=df, kind='kde')
plt.show()
```
